--> Solution for 4(b) <--

```python
DFAEven0 = md2mc('''DFA
IF_0Even : 0 -> 0Odd
0Odd : 0 -> IF_0Even
IF_0Even : 1 -> IF_0Even
0Odd : 1 -> 0Odd
''')
```

```python
dotObj_dfa(DFAEven0)
```

```python
DFAOdd1 = md2mc('''DFA
I_1Even : 1 -> F_1Odd
F_1Odd : 1 -> I_1Even
I_1Even : 0 -> I_1Even
F_1Odd : 0 -> F_1Odd
''')
```

```python
dotObj_dfa(DFAOdd1)
```

```python
dotObj_dfa(min_dfa(intersect_dfa(DFAEven0, DFAOdd1)))
```

```python
DFAOdd0OREven1Alt = comp_dfa(min_dfa(intersect_dfa(DFAEven0, DFAOdd1)))
```

```python
dotObj_dfa(DFAOdd0OREven1Alt)
```
